```python
import pandas as pd
```

```python
ohlc_dict = {
    'Open':'first',
    'High':'max',
    'Low':'min',
    'Close':'last',
    'Volume':'sum'
    }
df = pd.read_csv('bank_nifty1.csv')
df['Date Time'] = pd.to_datetime(df['Date Time'])
df.set_index('Date Time', inplace=True)
```

```python
df = df.resample('19Min', how=ohlc_dict).dropna(how='any') 
cols=['Open', 'High', 'Low', 'Close','Volume']  
df = df[cols]
df.head(100)
```

```python
df.to_csv("nifty5min.csv")
```

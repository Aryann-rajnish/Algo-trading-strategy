```python
data1 = pd.read_csv("bank_nifty.csv",index_col=0,parse_dates=[['Date', 'Time']])
data1.head()
```

```python
# changing into 5 minute time frame
def data5_min(data1):
    data_5min = pd.DataFrame(data1['Open'].resample('5Min').ohlc().open)
    data_5min['high'] = data1['High'].resample('5Min').ohlc().high
    data_5min['low'] = data1['Low'].resample('5Min').ohlc().low
    data_5min['close'] = data1['Close'].resample('5Min').ohlc().close
    
    return data_5min
```

```python
data_5min = data5_min(data1)
data_5min.head(80)
```

```python
data5 = data_5min.dropna(axis = 0, how ='any')
```

```python
data5.head(80)
```

```python
data5.to_csv("data5.csv")
```

```python
# changing into 15 minute time frame
def data15_min(data1):
    data_15min = pd.DataFrame(data1['Open'].resample('15Min').ohlc().open)
    data_15min['high'] = data1['High'].resample('15Min').ohlc().high
    data_15min['low'] = data1['Low'].resample('15Min').ohlc().low
    data_15min['close'] = data1['Close'].resample('15Min').ohlc().close
    return data_15min
```

```python
data15 = data15_min(data1)
data15 = data15.dropna(axis=0,how='any')
data15.to_csv("data15.csv")
```
